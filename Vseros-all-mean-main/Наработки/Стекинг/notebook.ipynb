```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'meta_feautres_data.csv')
```

```python
df = pd.read_csv(r'meta_feautres_data.csv')
df = df.pivot_table('target', 'index', 'model').reset_index().drop('index', axis=1)[['catboost', 'lgbm', 'xgb']]
lst = df.columns.tolist()
for i in lst:
    for u in lst:
        df[f'{i}_dif_{u}'] = df[i] - df[u]
        df[f'{i}_div_{u}'] = df[i] / df[u]
df = df.reset_index()
```
